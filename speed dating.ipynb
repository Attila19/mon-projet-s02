```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
dataset = pd.read_csv("Speed Dating Data.csv", encoding= "iso-8859-1")
dataset_01= dataset[["income","match"]]
```

```
dataset_01.groupby('income').count().sort_values(by=['match'],ascending=False)
datasetUtil = dataset_01.groupby('income').count().sort_values(by=['match'],ascending=False).iloc[0:50,:]
#print(datasetUtil)
```

```
plt.figure(figsize=(20,10))
plt.xlabel("nb rdv")
plt.ylabel("salaires")
plt.title("RDV_tarif√©")
plt.barh(datasetUtil.index,datasetUtil["match"])
plt.show()
```
